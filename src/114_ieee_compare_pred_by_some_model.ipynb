```python
import pandas as pd
```

```python
pred = pd.read_csv('../output/0826_ieee__clf_pred__high_fraud_user.csv')
pred_no_c = pd.read_csv('../output/0826_ieee__clf_pred__high_fraud_user__no_C.csv')
pred_no_c_card = pd.read_csv('../output/0826_ieee__clf_pred__high_fraud_user__no_C_card_amt.csv')

pred['diff_pred'] = pred['isFraud'] - pred['prediction']
pred_no_c['diff_pred_no_c'] = pred_no_c['isFraud'] - pred_no_c['prediction']
pred_no_c_card['diff_pred_no_c_card'] = pred_no_c_card['isFraud'] - pred_no_c_card['prediction']

pred_no_c.rename(columns={'prediction':'prediction_no_c'}, inplace=True)
pred_no_c_card.rename(columns={'prediction':'prediction_no_c_card'}, inplace=True)
df = pred.set_index('TransactionID').join(pred_no_c.set_index('TransactionID')[['diff_pred_no_c', 'prediction_no_c']]).join(pred_no_c_card.set_index('TransactionID')[['diff_pred_no_c_card', 'prediction_no_c_card']])
```

```python
new_user = pd.read_csv('../output/0826_transaction_new_user.csv').values.ravel()
```

```python
# pd.set_option('max_columns', 200)
df['diff_email'] = (df['P_emaildomain'] != df['R_emaildomain'])*1
df[df['diff_email']==1].sort_values(by='diff_pred_no_c_card'
                                    ,ascending=False
                                   )[sorted(
    [col for col in df.columns 
#      if  
#      not col.count('D')
#      and not col.count('C') 
#      and not col.count('D')
#      and not col.startswith('card')
#      and not col.count('addr')
#      and not col.count('M')
    ])]
```

```python
# df[['D1', 'D10', 'D11', 'C1', 'isFraud']]
df['D11'].fillna(-1, inplace=True)
df.groupby('D11')[['D1', 'C1', 'isFraud']].mean()
```

```python
df_new_user = pd.read_csv('../output/0826_ieee__same_user.csv')
df_max_fraud = df_new_user[df_new_user['max_Fraud']==1].set_index('TransactionID')
df_compare_same = df.join(df_max_fraud[['same_user_id_p', 'some_fraud_user']])
```
